```python
from option_valuation import monte_carlo, lattice
```

```python
S = 189.46
K = 50
r = 0.04581
vol = 0.2632
T = 194/365
j = 194
trials = 5000
call = True
dividends = ((0.23, 53/365),)
```

```python
mc_theo, var, paths = monte_carlo(S, K, r, vol, T, j, trials, call)
print("Theoretical value:", mc_theo,"with variance,", var)
```

```python
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import numpy as np
```

```python
x = np.arange(0, T+(1/365), T/j)

fig, ax = plt.subplots()
ax.set_xlim(0, T)
ax.set_ylim(paths.min(), paths.max())

segs = [np.column_stack([x, path]) for path in paths]

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

line_segments = LineCollection(segs, colors = colors, linestyle = 'solid')

ax.add_collection(line_segments)
ax.set_title('Stock Price Monte Carlo Paths')
plt.xlabel('Price')
plt.ylabel('Time')
plt.show()
```
